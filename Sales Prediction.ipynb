### HANDLING NULL VALUES IN DATAFRAME

```python
sales_prediction_data2 = sales_prediction_data.copy()
```

```python
null_series = sales_prediction_data2.isnull().sum()
null_series = null_series[null_series > 0]
null_series
```

```python
null_columns = list(null_series.index)
null_columns = sales_prediction_data2[null_columns]
```

```python
null_columns
```

```python
### Percentage of null values
null_columns_sum = null_columns.isnull().sum()
percentage_null = null_columns_sum/ 1460 * 100
```

```python
null_percent_df = pd.DataFrame(percentage_null, columns=["Percentage"])
null_percent_df
```

```python
null_columns.drop(["MiscFeature", "Fence", "PoolQC", "Alley"], axis=1, inplace=True)
```

```python
null_columns
```

```python
sales_prediction_data2.drop(["MiscFeature", "Fence", "PoolQC", "Alley"], axis=1, inplace=True)
```

```python
null_dtypes_dict = {str(k): list(v) for k, v in null_columns.groupby(null_columns.dtypes, axis=1)}
null_dtypes_dict
```

```python
sales_prediction_data2[null_dtypes_dict["float64"]].median()
```

```python
sales_prediction_data2[null_dtypes_dict["float64"]] = sales_prediction_data2[null_dtypes_dict["float64"]].fillna(sales_prediction_data2[null_dtypes_dict["float64"]].median())
```

```python
sales_prediction_data2[null_dtypes_dict["object"]] = sales_prediction_data2[null_dtypes_dict["object"]].fillna(sales_prediction_data2[null_dtypes_dict["object"]].max())
```

### HANDLING OUTLIERS 

```python
dict1 = {str(k): list(v) for k, v in sales_prediction_data2.groupby(sales_prediction_data2.dtypes, axis=1)}
dict1
```

```python
sales_prediction_num_column = sales_prediction_data2.drop(dict1["object"], axis=1)
```

### Finding Outliers using zscore

```python
from scipy import stats
z = np.abs(stats.zscore(sales_prediction_num_column))
z
```

### REMOVING OUTLIERS

```python
threshold = 3
filtered_entries = (z < 3).all(axis=1)
sales_prediction_data2[filtered_entries]
```

```python
sales_prediction_data2 = sales_prediction_data2[filtered_entries]
```

### Converting Categorical Columns in Numerical  

```python
object_columns = dict1["object"]
object_columns
```

```python
categorical_dummies = pd.get_dummies(sales_prediction_data2[object_columns])
```

```python
sales_prediction_data2
```

```python
sales_prediction_data2.drop(dict1["object"], axis=1, inplace= True)
```

```python
sales_prediction_data2
```

```python
sales_prediction_data2 = pd.concat([sales_prediction_data2, categorical_dummies], axis=1)
sales_prediction_data2
```

```python
sales_prediction_data2.set_index(pd.Index([i for i in range(len(sales_prediction_data2))]), inplace=True) 
sales_prediction_data2
```

```python
X = sales_prediction_data2.columns.values.tolist()
X.remove("SalePrice")
X = sales_prediction_data2[X]
Y = sales_prediction_data2["SalePrice"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


model = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8) 
```

```python
model.fit(x_train, y_train)
```

```python
sales_prediction = model.predict(x_test)
temp = x_test.copy()
```

```python
model.score(x_train,y_train)
```

```python
temp["Predicted Sales"] = sales_prediction
temp["Actual Sales"] = y_test
```

```python
temp["Predicted Sales"] = temp["Predicted Sales"].apply(lambda a: round(a))
```
